```python
#Import Important modules
import numpy as np
import pandas as pd
import ast

from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.stem.porter import PorterStemmer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Load the datatsets
credits_data = pd.read_csv("/content/credits.csv")
movies_data = pd.read_csv("/content/movies.csv")
```

```python
#Data preparation
data = movies_data.merge(credits_data, on = 'title')
```

```python
data.shape
```

```python
data.info()
```

```python
data = data[['movie_id', 'title', 'overview', 'genres','keywords', 'cast','crew']]
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace =True)
```

```python
data.isnull().sum()
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
data['genres'] = data['genres'].apply(convert)
data['keywords'] = data['keywords'].apply(convert)
```

```python
data.head()
```

```python
def convert3 (obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
data['cast'] = data['cast'].apply(convert3)
```

```python
data.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
data['crew'] = data['crew'].apply(fetch_director)
data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data
```

```python
data['genres'] = data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords'] = data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast'] = data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew'] = data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data
```

```python
data['tags'] = data['overview'] + data['genres']+ data['keywords'] + data['cast'] + data['crew']
```

```python
data
```

```python
new_data = data[['movie_id', 'title', 'tags']]
```

```python
new_data
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x: ' '.join(x))
```

```python
new_data
```

```python
new_data['tags'][0]
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data.head()
```

```python
#Features extraction
count = CountVectorizer(max_features = 5000, stop_words = 'english')
count.fit_transform(new_data['tags']).toarray().shape
```

```python
vectors = count.fit_transform(new_data['tags']).toarray()
vectors[0]
```

```python
len(count.get_feature_names_out())
porter = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(porter.stem(i))
    return " ".join(y)
new_data['tags'] = new_data['tags'].apply(stem)
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6]
```

```python
#Recommender
def recommend (movie):
    movie_index = new_data[new_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_data.iloc[i[0]].title)
```

```python
#Testing
recommend('Avatar')
```
